```python
import numpy as np
arr=np.array([1,3,5,7,9])
sum=arr[0]+arr[2]
diff=arr[3]-arr[2]
product=arr[1]*arr[3]
div=arr[4]/arr[2]
print("sum of 2 numbers=",sum)
print("difference of 2 numbers=",diff)
print("product of 2 numbers=",product)
print("division of 2 numbers=",div)
```

```python
#slicing of 1 dimension arrray
import numpy as np
arr=np.array([2,4,6,8,10,12,14,16])
print(arr[3:6])
print(arr[:5])
print(arr[2:])
print(arr[-5:-1])
```

```python
#element wise arithmetic operations
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
print(arr1@arr2)
```

```python
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
print()
concatenated_arr=np.concatenate((arr1,arr2),axis=0)
print(concatenated_arr)
print()
concatenated_arr=np.concatenate((arr1,arr2),axis=1)
print(concatenated_arr)
```
