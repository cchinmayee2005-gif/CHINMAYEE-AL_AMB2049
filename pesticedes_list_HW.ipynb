```python
pesticides=["neemoil","chloropyriphos","malathion","imidacloprid","spinosad"]
effectiveness=[75,85,70,90,88]
for p,e in zip(pesticides,effectiveness):
  print(p,"pesticide",e,"effectiveness")
```

```python
max_e = -1
best_pesticide = ""

for p, e in zip(pesticides, effectiveness):
  if e > max_e:
    max_e = e
    best_pesticide = p

print(f"The pesticide with the highest effectiveness is '{best_pesticide}' with an effectiveness of {max_e}.")
```
