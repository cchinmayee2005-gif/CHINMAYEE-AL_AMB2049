# extarcting all the strings with a prefix

```python
codes=[
 "CRP_RICE_2026",
 "CRP_WHEAT_2025",
 "FRT_MANGO_2026",
 "CRP_MAIZE_2024"
]
filtered=[]

for crop_codes in codes:
  if crop_codes.startswith("CRP"):#True
    filtered.append(crop_codes)

print(filtered)
```

```python
codes=[
  "CRP_RICE_2026",
  "CRP_WHEAT_2025",
  "FRT_MANGO_2026",
  "CRP_MAIZE_2024"
]
clean_codes=[]
for code in codes:
  if code.startswith("CRP_"):
    clean_codes.append(code[4:])
print(clean_codes)
```

# mini agricultural data cleaning example

```python
raw_record="rice,1200,black soil**"
clean_record=raw_record.strip()
clean_record=raw_record.strip("*")
crop,yield_kg,soil=clean_record.split(",")
crop=crop.title()

print(f"crop:{crop}")
print(f"yield:{yield_kg}kg")
print(f"soil type:{soil}")
```
