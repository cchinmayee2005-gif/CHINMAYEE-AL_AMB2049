```python
def evaluate_marks(marks):
    if marks >= 0 and marks <= 100:
        if marks >= 90:
            grade = "A+"
            message = "Excellent work!"
        elif marks >= 80:
            grade = "A"
            message = "Very good!"
        elif marks >= 70:
            grade = "B"
            message = "Good effort."
        elif marks >= 60:
            grade = "C"
            message = "Satisfactory."
        else:
            grade = "F"
            message = "Needs significant improvement."
        print(f"Marks: {marks}, Grade: {grade}, Message: {message}")
    else:
        print("Invalid marks. Marks should be between 0 and 100.")

# Test cases
evaluate_marks(95)
evaluate_marks(82)
evaluate_marks(70)
evaluate_marks(65)
evaluate_marks(45)
evaluate_marks(105)
evaluate_marks(-10)
```
