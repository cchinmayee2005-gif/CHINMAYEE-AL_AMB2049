```python
equipment=["tractor","plough","harvesting","seeder","sprayer","thresher"]
cost=[850000,25000,1200000,60000,18000,95000]
for p,e in zip(equipment,cost):
  max_cost=max(zip(cost,equipment))
print(max_cost)
min_cost=min(zip(cost,equipment))
print(min_cost)
```

```python
for i in range(len(equipment)/len(cost)):
  d=max_cost-min_cost
  print(" the max_cost and min_cost is,d ,%")
```
