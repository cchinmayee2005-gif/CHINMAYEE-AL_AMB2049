```python
import pandas as pd
production=pd.Series([30,60,45,25],
index=(["wheat","rice","maize","barley"]))
print(production)
```

```python
#bycrop name label
print(production["rice"])
print(production.loc["rice"])
```

```python
print(production.sum())
print(production.mean())
print(production.median())
print(production.idxmax())
```

```python
increased_production=production*1.10
print(increased_production)
```

```python
high_production=production[production>40]
print(high_production)
```

```python

import pandas as pd
data={
    "Farm":["Farm A","Farm C","Farm D","Farm B"],
    "Crop":["Wheat","Rice","maize","barley"],
    "Area_hectares":[10,15,8,12],
    "production_tons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.head(3)
```

```python
df.tail(2)
```

```python
df.info()
```

```python
df.describe()
```

how to add a new column to the existing one
#yield=production/area

```python
df["yield_ton_per_hectare"]=df["production_tons"]/df["Area_hectares"]
print(df)
```

```python
print(df.iloc[[1,3]])
```

```python
print(df.iloc[[1,3]])
```

```python
print(df.iloc[[1,3],[0,1,2]])
```

#filter data
#farms growing wheat

```python
print(df["Crop"]=="wheat")
print()
wheat_farms=df[df["Crop"]=="wheat"]
print(wheat_farms)
```

```python
high_yield=df[df["yield_ton_per_hectare"]>3]
print(high_yield)
```

```python
sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
avg_yield=df.groupby("Crop")["yield_ton_per_hectare"].sum()
print(avg_yield)
```
