```python
equipment=["tractor","plough","harvester","seeder","sprayer","thresher"]
cost=[850000,25000,1200000,60000,18000,95000]
updated_cost=[]
for y in cost:
  updated_cost.append(y*108),2
print("actual cost:",cost)
print("updated costs:",updated_cost)
```

```python
cost=[]
for c in updated_cost:
   if c<100000:
     cost.append("low")
   elif c<=500000:
     cost.append("medium")
   else:
    cost.append("high")
print("cost:",cost)
```

```python

subsidy_cost=[]
for c in cost:
  if c < 100000:
    subsidy_cost.append(int(c*0.85))
  else:
    subsidy_cost.append(c)
print("cost after subsidy:",subsidy_cost)
```

```python
subsidy_cost=[int(c*0.85) if c<100000 else c for c in cost]
print(subsidy_cost)
```
