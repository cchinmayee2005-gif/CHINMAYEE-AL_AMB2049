```python
import matplotlib.pyplot as plt
years=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
open_sensex=[
17473,20622,15535,19513,21222,27486,26102,26711,34060,36162,41349]
close_sensex=[20509,15455,19427,21171,27499,26118,26626,34057,36068,41254,33605]
plt.plot(years,open_sensex,'--',label='open',marker='o',markerfacecolor='red')
plt.plot(years,close_sensex,label='close')
plt.ylabel("sensex")
plt.xlabel("year")
plt.title("Sensex per year graph")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar Plot
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
open_sensex=[
17473,20622,15535,19513,21222,27486,26102,26711,34060,36162,41349]
close_sensex=[20509,15455,19427,21171,27499,26118,26626,34057,36068,41254,33605]
ind=np.arange(len(years))
print(ind)
width=0.3
plt.bar(ind,open_sensex,width,label="open")
plt.bar(ind+width,close_sensex,width,label="close",color='lavender')
plt.title("Sensex graph per year")
plt.ylabel("sensex")
plt.xlabel("year")
plt.xticks(ind, years,rotation=30)
plt.legend()
plt.show()
```

```python
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
open_sensex=[17473,20622,15535,19513,21222,27486,26102,26711,34060,36162,41349]
close_sensex=[20509,15455,19427,21171,27499,26118,26626,34057,36068,41254,33605]
ind=np.arange(len(years))
print(ind)
width=0.3
plt.barh(ind,open_sensex,width,label="open")
plt.barh(ind+width,close_sensex,width,label="close",color='g')
plt.title("Sensex graph per year")
plt.ylabel("sensex")
```
